```python
img7 = cv2.imread("7.jpg", cv2.IMREAD_COLOR)
img7_convert = cv2.cvtColor(img7, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img7)
cv2_imshow(img7_convert)
```

```python
img7_thresh1 = cv2.threshold(cv2.GaussianBlur(img7_convert, (5, 5), 0), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
cv2_imshow(img7_thresh1)
```

```python
img7_thresh2 = cv2.threshold(cv2.bilateralFilter(img7_convert, 5, 75, 75), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
cv2_imshow(img7_thresh2)
```

```python
img7_thresh3 = cv2.threshold(cv2.medianBlur(img7_convert, 3), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
cv2_imshow(img7_thresh3)
```

```python
img7_thresh4 = cv2.adaptiveThreshold(cv2.GaussianBlur(img7_convert, (5, 5), 0), 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 2)
cv2_imshow(img7_thresh4)
```

```python
img7_thresh5 = cv2.adaptiveThreshold(cv2.bilateralFilter(img7_convert, 9, 75, 75), 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 2)
cv2_imshow(img7_thresh5)
```

```python
img7_thresh6 = cv2.adaptiveThreshold(cv2.medianBlur(img7_convert, 3), 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 2)
cv2_imshow(img7_thresh6)
```
